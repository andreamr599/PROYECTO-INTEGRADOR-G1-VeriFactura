```python
from typing import Dict, Any

def build_output(doc_path: str, fields: Dict[str, Any], validations: Dict[str, Any], confidence: Dict[str, float]) -> Dict[str, Any]:
    return {
        "document": doc_path,
        "schema_version": "v1",
        "fields": fields,
        "validations": validations,
        "confidence": confidence
    }
```
