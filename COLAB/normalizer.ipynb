```python
import unicodedata, re
from typing import Dict

COMBUSTIBLE_MAP = {
    "EXTRA": "EXTRA",
    "DIESEL": "DIESEL",
    "GASOLINA": "GASOLINA",
    "GASOLINE": "GASOLINA",
    "GASOIL": "DIESEL"
}
```

```python
def strip_accents(s: str) -> str:
    if s is None: return s
    return ''.join(c for c in unicodedata.normalize('NFKD', s) if not unicodedata.combining(c))

def clean_text(s: str) -> str:
    if not s: return ""
    s = strip_accents(s)
    s = s.replace('\xa0', ' ')
    s = re.sub(r'\s+', ' ', s)
    return s.strip()
```

```python
def norm_money(s: str):
    if s is None: return None
    s = s.replace(' ', '')

    if ',' in s and '.' in s:
        if s.rfind(',') > s.rfind('.'):
            s = s.replace('.', '').replace(',', '.')
        else:
            s = s.replace(',', '')
    else:
        s = s.replace(',', '.')
    try:
        return float(re.sub(r'[^0-9.\-]', '', s))
    except:
        return None
```

```python
def norm_int(s: str):
    if s is None: return None
    s = re.sub(r'[^0-9]', '', s)
    return int(s) if s.isdigit() else None

def norm_combustible(s: str):
    if not s: return None
    s2 = clean_text(s).upper()
    for k,v in COMBUSTIBLE_MAP.items():
        if k in s2:
            return v
    return s2 or None
```
